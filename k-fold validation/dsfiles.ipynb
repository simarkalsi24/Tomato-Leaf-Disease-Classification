```python
import os

# Path to the original dataset folder containing all class folders
dataset_path = 'augmented_dataset'  # Change this to your actual path

# List all subdirectories (classes)
for class_name in os.listdir(dataset_path):
    class_path = os.path.join(dataset_path, class_name)
    
    if os.path.isdir(class_path):
        file_count = len([
            f for f in os.listdir(class_path)
            if os.path.isfile(os.path.join(class_path, f)) and not f.startswith('.')
        ])
        print(f"{class_name}: {file_count} image files")
```

```python
import os

def review_dataset_structure(root_dir):
    for dirpath, dirnames, filenames in os.walk(root_dir):
        print(f"\n📁 Folder: {dirpath}")
        if dirnames:
            print(f"  ├─ Subfolders: {', '.join(dirnames)}")
        if filenames:
            print(f"  ├─ Files:")
            for file in filenames[:10]:  # show only first 10
                print(f"      - {file}")

# 🔧 Set your dataset path
dataset_path = "augmented_dataset"  # replace with your actual path
review_dataset_structure(dataset_path)
```

```python
import os

# Set the path to your dataset folder
dataset_path = 'augmented_dataset'  # <-- change this

# Counter to keep track of removed files
deleted_count = 0

# Walk through all subdirectories
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file == '.DS_Store':
            file_path = os.path.join(root, file)
            try:
                os.remove(file_path)
                deleted_count += 1
                print(f"Deleted: {file_path}")
            except Exception as e:
                print(f"Error deleting {file_path}: {e}")

print(f"\n✅ Total .DS_Store files deleted: {deleted_count}")
```

```python
import cv2
dataset_path = 'augmented_dataset'
def is_blurry(image_path, threshold=100.0):
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None: return True
        return cv2.Laplacian(img, cv2.CV_64F).var() < threshold
    except:
        return True

def remove_blurry_images(folder_path):
    for subdir, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                path = os.path.join(subdir, file)
                if is_blurry(path):
                    os.remove(path)
                    print(f"Removed blurry image: {path}")

remove_blurry_images(dataset_path)
```
